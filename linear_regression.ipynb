```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "size":[1000,1500,2000,2500],
    "price":[200000,300000,400000,500000]
}

df = pd.DataFrame(data)
```

```python
X = df[['size']]   # input
y = df['price']    # output
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
prc = int(input("Enter the size of house: "))

price = model.predict([[prc]])

print("Predicted Price:", price)
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X))
plt.xlabel("House Size")
plt.ylabel("Price")
plt.show()
```

```python
# this is LinearRegression
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {
    "size":[1000,1500,2000,2500],
    "price":[200000,300000,400000,500000]
}

df = pd.DataFrame(data)

X = df[['size']]
y = df['price']

model = LinearRegression()
model.fit(X,y)

size = int(input("Enter the size of house: "))

new_house = pd.DataFrame([[size]], columns=['size'])

price = model.predict(new_house)

print("Predicted Price:", price)
```

```python

# DecisionTreeClassifire
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Dataset
data = {
    "Age":[20,22,25,30,35,40],
    "Gender":[1,0,1,0,1,0],  # Male=1 Female=0
    "Genre":[0,1,0,2,3,2]
}

df = pd.DataFrame(data)

# Features and target
X = df[["Age","Gender"]]
y = df["Genre"]

# Train Decision Tree
model = DecisionTreeClassifier()
model.fit(X,y)

# User input
age = int(input("Enter Age: "))

# Option 1: Ask user to enter 0/1
gender = int(input("Enter Gender (Male=1, Female=0): "))

row = [age, gender]
new_data = pd.DataFrame([row], columns=['Age','Gender'])

# Convert to dataframe to match training format
# new_data = pd.DataFrame([[age, gender]], columns=['Age','Gender'])

# Prediction
prediction = model.predict(new_data)

# Map numeric genre to names
genres = ["Action","Romance","Drama","Thriller"]
print("Recommended Genre:", genres[prediction[0]])
```

```python
#RandomForestClassifier

import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Dataset
data = {
    "Age":[20,22,25,30,35,40],
    "Gender":[1,0,1,0,1,0],  # Male=1 Female=0
    "Genre":[0,1,0,2,3,2]
}

df = pd.DataFrame(data)

# Features and target
X = df[['Age','Gender']]
y = df['Genre']

# Create Random Forest model
model = RandomForestClassifier(n_estimators=5, random_state=42)  # 5 trees

# Train model
model.fit(X,y)

# Input from user
age = int(input("Enter Age: "))
gender_input = input("Enter Gender (M/F): ").upper()
gender = 1 if gender_input == 'M' else 0

# Prepare input as DataFrame
new_data = pd.DataFrame([[age, gender]], columns=['Age','Gender'])

# Predict
prediction = model.predict(new_data)

genres = ["Action","Romance","Drama","Thriller"]
print("Recommended Genre:", genres[prediction[0]])
```
