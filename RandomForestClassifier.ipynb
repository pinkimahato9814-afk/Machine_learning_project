```python
#RandomForestClassifier

import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Dataset
data = {
    "Age":[20,22,25,30,35,40],
    "Gender":[1,0,1,0,1,0],  # Male=1 Female=0
    "Genre":[0,1,0,2,3,2]
}

df = pd.DataFrame(data)

# Features and target
X = df[['Age','Gender']]
y = df['Genre']

# Create Random Forest model
model = RandomForestClassifier(n_estimators=5, random_state=42)  # 5 trees

# Train model
model.fit(X,y)

# Input from user
age = int(input("Enter Age: "))
gender_input = input("Enter Gender (M/F): ").upper()
gender = 1 if gender_input == 'M' else 0

# Prepare input as DataFrame
new_data = pd.DataFrame([[age, gender]], columns=['Age','Gender'])
```
